```
dd.columns = ['Index', 'Student_name', 'University_Serial_No', 'Discipline', 'Year_of_Passing', 'Campus_type', 'Employer_Name', 'Status', 'Salary_LPA']
```

```
dd.head()
```

```
uni_disciplines=dd['Discipline'].unique()
uni_disciplines
```

```
dd['Salary_LPA']=pd.to_numeric(dd['Salary_LPA'],errors='coerce')
```

```
campus_type_counts = dd['Campus_type'].value_counts()
campus_type_counts
```

```
emp_name=dd['Employer_Name'].unique()
emp_name
```

```
avg_salary_per_discipline = dd.groupby('Discipline')['Salary_LPA'].mean()
avg_salary_per_discipline
```

```
offers_per_employer = dd.groupby('Employer_Name').size()
offers_per_employer
```

```
high_salary=dd[dd['Salary_LPA']>3]
high_salary
```

```
low_salary=dd[dd['Salary_LPA']<2]
low_salary
```

```
import matplotlib.pyplot as plt
placement_counts = dd['Discipline'].value_counts()
plt.bar(placement_counts.index, placement_counts.values)
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.title('Placement Counts')
plt.show()
```

```
type_placement_counts = dd['Campus_type'].value_counts()
plt.bar(type_placement_counts.index, type_placement_counts.values)
plt.xlabel('Campus_type')
plt.ylabel('Count')
plt.title('Placement Counts')
plt.show()
```

```
Discipline_counts = dd['Discipline'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(Discipline_counts, labels=Discipline_counts.index, autopct='%.4f')
plt.show()
```
